```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/data mining 11.11.2025/shopping_behavior_updated.csv")
data.info()
```

```python
data.isnull().sum()
```

```python
import seaborn as sns
sns.set(style="whitegrid")
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Age", y="Purchase Amount (USD)", data=data, ci=None)
plt.title("Average Purchase Amount by Age")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y="Purchase Amount (USD)", data=data)
plt.title("Box Plot of Purchase Amount")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Age", y="Purchase Amount (USD)", data=data, marker="o")
plt.title("Trend of Purchase Amount with Age")
plt.xlabel("Age")
plt.ylabel("Purchase Amount (USD)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Purchase Amount (USD)", data=data)
plt.title("Age vs Purchase Amount")
plt.xlabel("Age")
plt.ylabel("Purchase Amount (USD)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Review Rating", data=data, palette="pastel")
plt.title("Count of Each Review Rating")
plt.xlabel("Review Rating")
plt.ylabel("Count")
plt.show()
```

```python
corr = data[['Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, linecolor='black')
plt.title("Correlation Heatmap of Shopping Behavior Data")
plt.show()
```

```python
corr = data[['Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases']].corr()

plt.figure(figsize=(9,7))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="Spectral",
    center=0,
    linewidths=1,
    linecolor='black',
    annot_kws={"size":14, "weight":"bold", "color":"black"}
)

plt.title("Enhanced Correlation Heatmap of Shopping Behavior Data", fontsize=16, weight='bold')
plt.xticks(fontsize=12, weight='bold')
plt.yticks(fontsize=12, weight='bold')
plt.show()
```
